```python
for i in df:
  if df[i].dtype=='int' or df[i].dtype=='float':
    print(df[i].value_counts())
    print('_________________________________________________')
```

```python
object_type=[]    #creating empty list
for i in df:      #iterating through the df
  if df[i].dtype=='object': #if the datatype is object
    object_type.append(i)
```

```python
object_type
```

```python
from sklearn.preprocessing import LabelEncoder
dict1={}                  #creating empty dict
for i in object_type:     #iterating through the list of columns
  dict1[i]=LabelEncoder() #creating seperate objects for each column
  df[i]=dict1[i].fit_transform(df[i])
```

```python
x=df.drop(columns=['Species'])
y=df['Species']
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler  #calling MinMaxScaler and StandardScaler class from sklearn.preprocessing
minmax=MinMaxScaler()                           #creating an object
x_scaled=minmax.fit_transform(x)
```

```python
scaler=StandardScaler()
x_scaled_new=scaler.fit_transform(x)
x_scaled_new.max()
x_scaled_new.min()
```

```python
from sklearn.model_selection import train_test_split    #train_test_split is a function and not a class
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.3,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier      #importing kneighbour classifier from sklearn
model=KNeighborsClassifier()                            #creating an object to train using knc, there are many parameters in this one of them is n neighbours which is by default set to 5
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_test.values
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score    #importing accuracy score function from sklearn library which has metrics module
print(accuracy_score(y_test,y_pred))    #gives the accuracy by comparing y_test(actual correct values) with y_pred(predicted values), each value in y_test is compared to the corresponding value in y_pred
print(f"Accuracy:{round(accuracy_score(y_test,y_pred)*100)}")
```

```python
from sklearn.svm import SVC #from sklearn library svm module we import svc(support vector classifier) class\
model=SVC()   #creating an obj
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(accuracy_score(y_test,y_pred))
model.score(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix    #importing confusion matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report   #import classification_report from sklearn metrics
print(classification_report(y_test,y_pred))
```
